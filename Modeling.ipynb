```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('./Telecom_customer churn cleaned.csv')
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df.head()
```

```python
df = df.drop(['Customer_ID'], axis = 1)
```

```python
df_dum = pd.get_dummies(df)
df_dum.head()
```

```python
X = df_dum.drop('churn', axis = 1) 
y = df_dum.churn
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
from sklearn.metrics import accuracy_score
```

**Here I run Random Forest Classification after which I check the important feautures. After choosing important feautures, I feed the next classifiers with only those important features.**

```python
# Fitting Random Forest Classification to the Training set
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 42)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(round(accuracy_score(y_test,y_pred) * 100, 2))
```

```python
importances=pd.Series(classifier.feature_importances_, index=X.columns)
importances.plot(kind='barh', figsize=(30,100))
```

```python
sorted_indices = np.argsort(classifier.feature_importances_)[::-1]

sorted_indices
```

```python
# Fitting Random Forest Classification to the Training set
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 42)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(round(accuracy_score(y_test,y_pred) * 100, 2))
```

```python
# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

print(round(accuracy_score(y_test,y_pred) * 100, 2))
```

```python
# Fitting Naive Bayes to the Training set
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(round(accuracy_score(y_test,y_pred) * 100, 2))
```
