```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('./Telecom_customer churn cleaned.csv')
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.describe()
```

```python
cmap = sns.diverging_palette(220, 10, as_cmap=True)

fig, ax = plt.subplots(figsize=(15,15))

yticks = df.index
keptticks = yticks[::int(len(yticks)/10)]
yticks = ['' for y in yticks]
yticks[::int(len(yticks)/10)] = keptticks

xticks = df.columns
keptticks = xticks[::int(len(xticks)/10)]
xticks = ['' for y in xticks]
xticks[::int(len(xticks)/10)] = keptticks

sns.heatmap(df.corr(), cmap="YlGnBu", vmax=1, annot=False, center = 0, square = True, 
            linewidths=.5, ax=ax, xticklabels=True, yticklabels=True, cbar_kws={"shrink": .5})

plt.xticks(rotation=90) 

plt.show()
```
